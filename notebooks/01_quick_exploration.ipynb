# RA-PainKG: Quick Exploration Notebook

Load and explore the tissue-specific knowledge graph for RA pain signaling.

```python
import sys
sys.path.insert(0, "../src")

import networkx as nx
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import json
from pathlib import Path

%matplotlib inline
plt.rcParams["figure.dpi"] = 100
```

## 1. Load the Knowledge Graph

```python
# Load from pickle (fastest)
with open("../output/RA_PainKG_final.pkl", "rb") as f:
    g = pickle.load(f)

print(f"Nodes: {g.number_of_nodes():,}")
print(f"Edges: {g.number_of_edges():,}")
print(f"Directed: {g.is_directed()}")
```

## 2. Node Type Distribution

```python
from collections import Counter
node_types = Counter(d.get("node_type", "unknown") for _, d in g.nodes(data=True))
pd.DataFrame(node_types.most_common(), columns=["Node Type", "Count"])
```

## 3. Track Gene Exploration

```python
# Get Track A and Track B genes
track_a_genes = [(n, d["node_name"]) for n, d in g.nodes(data=True) if d.get("track_a")]
track_b_genes = [(n, d["node_name"]) for n, d in g.nodes(data=True) if d.get("track_b")]
dual_genes = [(n, d["node_name"]) for n, d in g.nodes(data=True) if d.get("track_a") and d.get("track_b")]

print(f"Track A (Inflammation): {len(track_a_genes)} genes")
print(f"Track B (Nociception): {len(track_b_genes)} genes")
print(f"Dual Track: {len(dual_genes)} genes")
print(f"\nSample Track B genes: {[g[1] for g in track_b_genes[:10]]}")
```

## 4. Top Hub Nodes

```python
# Load precomputed centrality
centrality = pd.read_csv("../output/gene_centrality.csv")
top_10 = centrality.nlargest(10, "betweenness")[["node_name", "betweenness", "pagerank", "track_a", "track_b"]]
top_10
```

## 5. Query a Specific Gene

```python
def query_gene(g, gene_symbol):
    """Find a gene node and its neighborhood."""
    for node, data in g.nodes(data=True):
        if str(data.get("node_name", "")).upper() == gene_symbol.upper():
            print(f"Found: {data['node_name']}")
            print(f"  Type: {data.get('node_type')}")
            print(f"  Track: A={data.get('track_a')}, B={data.get('track_b')}")
            print(f"  Degree: {g.degree(node)}")
            print(f"\nNeighbors:")
            for neighbor in list(g.successors(node))[:10]:
                nd = g.nodes[neighbor]
                print(f"  -> {nd.get('node_name', neighbor)} ({nd.get('node_type', '?')})")
            return
    print(f"Gene {gene_symbol} not found in KG")

query_gene(g, "TRPV1")
```

## 6. Track A vs Track B Topology Comparison

```python
import numpy as np

track_a_bc = centrality[centrality["track_a"]]["betweenness"]
track_b_bc = centrality[centrality["track_b"]]["betweenness"]

fig, ax = plt.subplots(figsize=(8, 5))
ax.boxplot([track_a_bc, track_b_bc], labels=["Track A\n(Inflammation)", "Track B\n(Nociception)"])
ax.set_ylabel("Betweenness Centrality")
ax.set_title("Track A vs Track B: Betweenness Centrality")
plt.tight_layout()
plt.show()
```
